```python
print('check the nan value in labels_DataFrame')
labels_DataFrame.isnull().sum()
```

```python
def missingdata(labels_DataFrame):
    total = labels_DataFrame.isnull().sum().sort_values(ascending = False)
    percent = (labels_DataFrame.isnull().sum()/labels_DataFrame.isnull().count()*100).sort_values(ascending = False)
    ms=pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    ms= ms[ms["Percent"] > 0]
    f,ax =plt.subplots(figsize=(8,6))
    plt.xticks(rotation='90')
    fig=sns.barplot(ms.index, ms["Percent"],color="green",alpha=0.8)
    plt.xlabel('Labels', fontsize=15)
    plt.ylabel('Percent of missing values', fontsize=15)
    plt.title('Percent missing data by feature', fontsize=15)
    return ms
missingdata(labels_DataFrame)
```

```python
labels_DataFrame['SR-MMP'].fillna(labels_DataFrame['SR-MMP'].mode()[0], inplace = True)
labels_DataFrame['NR-Aromatase'].fillna(labels_DataFrame['NR-Aromatase'].mode()[0], inplace = True)
labels_DataFrame['SR-ARE'].fillna(labels_DataFrame['SR-ARE'].mode()[0], inplace = True)
labels_DataFrame['NR-ER'].fillna(labels_DataFrame['NR-ER'].mode()[0], inplace = True)
labels_DataFrame['NR-PPAR-gamma'].fillna(labels_DataFrame['NR-PPAR-gamma'].mode()[0], inplace = True)
labels_DataFrame['SR-HSE'].fillna(labels_DataFrame['SR-HSE'].mode()[0], inplace = True)
labels_DataFrame['NR-AhR'].fillna(labels_DataFrame['NR-AhR'].mode()[0], inplace = True)
labels_DataFrame['NR-AR-LBD'].fillna(labels_DataFrame['NR-AR-LBD'].mode()[0], inplace = True)
labels_DataFrame['SR-p53'].fillna(labels_DataFrame['SR-p53'].mode()[0], inplace = True)
labels_DataFrame['NR-ER-LBD'].fillna(labels_DataFrame['NR-ER-LBD'].mode()[0], inplace = True)
labels_DataFrame['SR-ATAD5'].fillna(labels_DataFrame['SR-ATAD5'].mode()[0], inplace = True)
labels_DataFrame['NR-AR'].fillna(labels_DataFrame['NR-AR'].mode()[0], inplace = True)
```
